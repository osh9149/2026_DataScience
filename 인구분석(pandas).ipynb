```python
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter("ignore")

df = pd.read_excel('2023_age.xlsx')
df
```

```python
df.info()
```

```python
df.dtypes
```

```python
df.astype({'총 인구수':int})
df.dtypes
```

```python
import matplotlib.ticker as mticker

df.plot(kind='bar',x='행정기관',y='총 인구수',color='red', figsize=(10,5), grid=True)
plt.title('2023 전국 총인구수')
plt.gca().yaxis.set_major_formatter(mticker.FormatStrFormatter('%10d 명'))
plt.savefig('test',dpi=300)
plt.show()
```

```python
df_age = df.transpose()
df_age
```

```python
df_age.rename(columns=df_age.iloc[1],inplace=True)

df_age
```

```python
df_age = df_age.drop(df_age.index[0:4])
df_age
```

```python
df_age.reset_index(drop=False, inplace=True)
df_age
```

```python
#column name을 index->  연령별로 변경

df_age.rename(columns={'index':'연령별'}, inplace=True)
df_age
```

```python
df_age.dtypes
```

```python
df_age.set_index('연령별',inplace=True)
df_age = df_age.astype('int')   #inplace 가 없음
df_age.reset_index(inplace=True)
df_age
```

```python

df_age.dtypes
```

```python
df_age.plot(kind='bar',x='연령별',y='전국', color='blue', figsize=(20,5), grid=True)
plt.title('2023 전국 연령별 인구수')
plt.show()
```

```python
df_age.plot(kind='barh',x='연령별',y='전국', color='green', figsize=(10,20), grid=True)
plt.title('2023 전국 연령별 인구수')
plt.show()
```
