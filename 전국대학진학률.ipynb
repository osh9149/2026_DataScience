3. 연도를 입력받아 진학률 조회

```python
import csv
f=open('jinhak2020.csv',encoding='cp949')
rows=csv.reader(f)
next(f)

max_num=0
max_loc=''
min_num=100
min_loc=''

year=input('연도입력')
# print(year)
i=0

for row in rows:
    if row[0]==year:  #연도별

      if max_num<float(row[5]):
        max_num=float(row[5])
        max_loc=row[1]+row[2]

      if min_num>float(row[5]):
        min_num=float(row[5])
        min_loc=row[1]+row[2]

print(max_num,max_loc)
print(min_num,min_loc)
```

4. 2010년부터 2020년까지 전국 대학진학률(최고, 최저 지역분석)

```python
import csv

year=list(range(2010,2021))
# print(year)
i=0
for yy in year:
  max_num=0
  min_num=100
  max_loc='#'
  min_loc='#'
  print('************'+str(yy)+'******************')
  f=open('jinhak2020.csv',encoding='cp949')
  rows=csv.reader(f)
  next(f)

  for row in rows:
    # print(row)
    if row[0]==str(yy):
      if max_num<float(row[5]):
        max_num=float(row[5])
        max_loc=row[1]+row[2]

      if min_num>float(row[5]):
        min_num=float(row[5])
        min_loc=row[1]+row[2]

  print(yy, max_num,max_loc)
  print(yy, min_num,min_loc)
```
