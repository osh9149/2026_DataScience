#2. 폴더에서 읽어오기(/content)

##csv 파일의 경우

```python
import pandas as pd

# 여기에 실제 CSV 파일 이름을 입력해주세요. (예: 'data.csv')
file_path = 'seoul_temp_c.csv'

try:
    df_local = pd.read_csv(file_path, encoding='cp949')
    print(f"'{file_path}' 파일을 성공적으로 불러왔습니다.")
    display(df_local.head())
except FileNotFoundError:
    print(f"오류: '{file_path}' 파일을 찾을 수 없습니다. 파일 이름을 확인하거나, 파일이 올바른 위치에 있는지 확인해주세요.")
except Exception as e:
    print(f"파일을 불러오는 중 오류가 발생했습니다: {e}")
```

##exel 파일의 경우(.xlsx)

```python
import pandas as pd

# 여기에 실제 Excel 파일 이름을 입력해주세요. (예: 'my_excel_data.xlsx')
file_path = 'seoul_temp.xlsx'

try:
    df_local_excel = pd.read_excel(file_path)
    print(f"'{file_path}' 파일을 성공적으로 불러왔습니다.")
    display(df_local_excel.head())
except FileNotFoundError:
    print(f"오류: 로컬에 '{file_path}' 파일을 찾을 수 없습니다. 파일 이름을 확인하거나, 파일이 올바른 위치에 있는지 확인해주세요.")
except Exception as e:
    print(f"파일을 불러오는 중 오류가 발생했습니다: {e}")
```
